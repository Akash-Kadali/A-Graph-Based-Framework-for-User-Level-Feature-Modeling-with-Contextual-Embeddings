```python
import pandas as pd

# Assuming language_spam_counts is the Series containing language counts

def split_by_language(data, language_counts, destination):
  dfs = []
  for language, count in language_counts.items():
    # Filter data for the current language
    df_filtered = data[data['lang'] == language]
    # Ensure there's data for the language (optional)
    if not df_filtered.empty:
      dfs.append(df_filtered)
  return dfs

# Assuming your original data is stored in 'mega_spam_user_tweets' DataFrame
language_dfs = split_by_language(mega_spam_user_tweets.copy(), language_spam_counts, "/content/drive/MyDrive/MiB Dataset/")

# Save each DataFrame as a CSV file in the specified destination
for i, df in enumerate(language_dfs):
  language = df['lang'].iloc[0]  # Assuming unique language per DataFrame
  filename = f'/content/drive/MyDrive/MiB Dataset/language_{language}.csv'  # Construct full path directly
  df.to_csv(filename, index=False)  # Avoid including index in CSV
```

```python
# Assuming your original data is stored in 'mega_spam_user_tweets' DataFrame
language_dfs = split_by_language(mega_real_user_tweets.copy(), language_real_counts, "/content/drive/MyDrive/MiB Dataset/User Languages/Real")

# Save each DataFrame as a CSV file in the specified destination
for i, df in enumerate(language_dfs):
  language = df['lang'].iloc[0]  # Assuming unique language per DataFrame
  filename = f'/content/drive/MyDrive/MiB Dataset/User Languages/Real/language_{language}.csv'  # Construct full path directly
  df.to_csv(filename, index=False)  # Avoid including index in CSV
```
