```python
import tensorflow as tf
from tqdm import tqdm

# Initialize TPU strategy
resolver = tf.distribute.cluster_resolver.TPUClusterResolver(tpu='')
tf.config.experimental_connect_to_cluster(resolver)
tf.tpu.experimental.initialize_tpu_system(resolver)
strategy = tf.distribute.TPUStrategy(resolver)

# Define your code within the strategy's scope
with strategy.scope():
    max_sampled_length = 0
    optimal_x = 0

    # Iterate over values of x
    for x in tqdm(range(1, 3500)):
        users_more_than_x = user_id_counts[user_id_counts > x].index.tolist()

        # Filter the original DataFrame based on the selected user IDs
        df_filtered = df[df['user_id'].isin(users_more_than_x)]

        # Group the DataFrame by user_id and select the first x tweets from each group
        df_sampled = df_filtered.groupby('user_id').head(x)

        # Calculate the length of the sampled DataFrame
        sampled_length = len(df_sampled)

        # Update max_sampled_length and optimal_x if a new maximum is found
        if sampled_length > max_sampled_length:
            max_sampled_length = sampled_length
            optimal_x = x

print("Optimal x:", optimal_x)
print("Highest length of df_sampled:", max_sampled_length)
```

```python
import pandas as pd
from tqdm import tqdm

social_spam_ratio = None

def calculate_ratio(n):
    global social_spam_ratio
    user_counts = df['user_id'].value_counts()
    valid_users = user_counts[user_counts >= n].index

    # Sample exactly n tweets for each valid user
    sampled_df = df[df['user_id'].isin(valid_users)].groupby('user_id').apply(lambda x: x.sample(n=n, random_state=42))  # Adjust random_state as needed

    # Assuming df is your DataFrame with the columns and sampled_df is the DataFrame where you want to append the columns
    columns_to_append = pd.DataFrame(columns=[
        'index', 'user_id', 'statuses_count', 'followers_count',
        'friends_count', 'favourites_count', 'listed_count', 'url',
        'lang', 'description', 'text', 'retweet_count', 'reply_count',
        'num_hashtags', 'num_urls', 'type', 'sample_type'
    ])

    sampled_df = pd.concat([columns_to_append, sampled_df], ignore_index=True)

    # Count occurrences of each type
    type_counts = sampled_df['type'].value_counts()

    # Calculate ratio
    social_spam_ratio = type_counts['social_spam'] / type_counts['real']


# Iterate over values of n from 1 to 200 and calculate ratio
nearest_to_one = float('inf')
nearest_n = None
for n in tqdm(range(1, 201)):  # Use tqdm to track progress
    calculate_ratio(n)
    ratio = abs(social_spam_ratio - 1)
    if ratio < nearest_to_one:
        nearest_to_one = ratio
        nearest_n = n

print("Nearest value of n for which the ratio is closest to 1:", nearest_n)
```

```python
import pandas as pd
from tqdm import tqdm

def calculate_social_spam_count(n):
    user_counts = df['user_id'].value_counts()
    valid_users = user_counts[user_counts >= n].index

    # Sample exactly n tweets for each valid user
    sampled_df = df[df['user_id'].isin(valid_users)].groupby('user_id').apply(lambda x: x.sample(n=n, random_state=42))  # Adjust random_state as needed

    # Assuming df is your DataFrame with the columns and sampled_df is the DataFrame where you want to append the columns
    columns_to_append = pd.DataFrame(columns=[
        'index', 'user_id', 'statuses_count', 'followers_count',
        'friends_count', 'favourites_count', 'listed_count', 'url',
        'lang', 'description', 'text', 'retweet_count', 'reply_count',
        'num_hashtags', 'num_urls', 'type', 'sample_type'
    ])

    sampled_df = pd.concat([columns_to_append, sampled_df], ignore_index=True)

    # Count occurrences of social spam
    social_spam_count = (sampled_df['type'] == 'social_spam').sum()

    return social_spam_count

# Iterate over values of n from 1 to 200 and find the value of n for which social_spam samples are maximum
max_social_spam_count = 0
max_n = None
for n in tqdm(range(1, 201), desc="Calculating"):
    social_spam_count = calculate_social_spam_count(n)
    if social_spam_count > max_social_spam_count:
        max_social_spam_count = social_spam_count
        max_n = n

print("For what value of n is the number of samples of social_spam maximum:", max_n)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# Precompute user counts
user_counts = df['user_id'].value_counts()

# Preallocate arrays for faster appending
n_values = range(50, 3200)
total_samples = np.zeros(len(n_values))
social_spam_counts = np.zeros(len(n_values))
real_counts = np.zeros(len(n_values))

# Filter the DataFrame once
valid_users = user_counts[user_counts >= max(n_values)].index
sampled_df = df[df['user_id'].isin(valid_users)].copy()

for i, n in enumerate(tqdm(n_values)):
    # Sample without replacement
    sampled_subset = sampled_df.groupby('user_id').apply(lambda x: x.sample(n=n, random_state=42))

    # Count types using value_counts
    counts = sampled_subset.groupby('type').size()
    total_samples[i] = counts.sum()
    social_spam_counts[i] = counts.get('social_spam', 0)
    real_counts[i] = counts.get('real', 0)

# Plot the line graph
plt.plot(n_values, total_samples, label='Total Samples')
plt.plot(n_values, social_spam_counts, label='Social Spam')
plt.plot(n_values, real_counts, label='Real')
plt.xlabel('n Value')
plt.ylabel('Count')
plt.title('Counts of Total Samples, Social Spam, and Real Samples for Different n Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from tqdm import tqdm

# Calculate the counts of each user ID
user_counts = df['user_id'].value_counts()

# Initialize lists to store counts for each value of n
total_samples = []
social_spam_counts = []
real_counts = []

# Specify the values of n
n_values = [2, 4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048]

# Iterate over values of n
for n in tqdm(n_values):
    # Filter the DataFrame to include only user IDs that occur at least n times
    valid_users = user_counts[user_counts >= n].index
    sampled_df = df[df['user_id'].isin(valid_users)].groupby('user_id').apply(lambda x: x.sample(n=n, random_state=42))

    # Calculate counts of social_spam and real samples
    total_sample_count = len(sampled_df)
    social_spam_count = sampled_df[sampled_df['type'] == 'social_spam'].shape[0]
    real_count = sampled_df[sampled_df['type'] == 'real'].shape[0]

    # Append counts to respective lists
    total_samples.append(total_sample_count)
    social_spam_counts.append(social_spam_count)
    real_counts.append(real_count)

# Plot the line graph
plt.plot(n_values, total_samples, label='Total Samples')
plt.plot(n_values, social_spam_counts, label='Social Spam')
plt.plot(n_values, real_counts, label='Real')
plt.xlabel('n Value')
plt.ylabel('Count')
plt.title('Counts of Total Samples, Social Spam, and Real Samples for Different n Values')
plt.xscale('log')  # Set x-axis to log scale
plt.legend()
plt.grid(True)
plt.show()
```
