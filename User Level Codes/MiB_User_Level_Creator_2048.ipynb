```python
import pandas as pd

# Assuming df_train is your DataFrame with tweets and user_id columns
# Step 1: Count the number of tweets for each user_id
user_id_counts = df_train['user_id'].value_counts()

# Step 2: Filter user_ids with 1024 tweets or fewer
filtered_user_ids = user_id_counts[user_id_counts <= 2048].index

# Step 3: Create a new DataFrame with only the filtered user_ids
filtered_df = df_train[df_train['user_id'].isin(filtered_user_ids)]

filtered_df["type"].value_counts()
```

```python
import pandas as pd
import numpy as np

# Step 1: Count the number of tweets for each user_id
user_id_counts = filtered_df['user_id'].value_counts()

# Step 2: Determine the maximum number of tweets any user has
max_tweets = 2048

# Step 3: Create a function to pad each user's data with zero rows
def pad_user_tweets(group, max_tweets):
    num_current_tweets = len(group)
    num_padding_tweets = max_tweets - num_current_tweets

    # Create a DataFrame with zero values for padding and matching "type"
    padding_df = pd.DataFrame({
        'user_id': [group['user_id'].iloc[0]] * num_padding_tweets,
        'text': [0] * num_padding_tweets,  # Assuming the text content column is named 'text'
        'type': [group['type'].iloc[0]] * num_padding_tweets,
        'sample_type': [group['sample_type'].iloc[0]] * num_padding_tweets,
        'lang': [group['lang'].iloc[0]] * num_padding_tweets
    })

    # Concatenate the original group with the padding DataFrame
    padded_group = pd.concat([group, padding_df], ignore_index=True)

    return padded_group

# Step 4: Apply the padding function to each user group
padded_df = filtered_df.groupby('user_id').apply(lambda group: pad_user_tweets(group, max_tweets)).reset_index(drop=True)
padded_df
```

```python
# Assuming padded_df is your DataFrame with potential NaN values
padded_df_filled = padded_df.fillna(0)
padded_df_filled
```

```python
padded_df_filled["user_id"].value_counts()
```

```python
# Step 4: Sort the filtered DataFrame by user_id
padded_df_filled_sorted = padded_df_filled.sort_values(by='user_id')
padded_df_filled_sorted
```

```python
# Function to shuffle each user's tweets
def shuffle_tweets(group):
    return group.sample(frac=1, random_state=42).reset_index(drop=True)

# Group by user_id, shuffle each group, and sort by user_id
new_df = padded_df_filled_sorted.groupby('user_id').apply(shuffle_tweets).reset_index(drop=True).sort_values(by='user_id').reset_index(drop=True)
new_df
```

```python
import pandas as pd
import numpy as np
import torch
from transformers import AutoTokenizer, AutoModel

# Load data
df = pd.read_csv("/content/drive/MyDrive/MiB Dataset/2048 Each User/Splits/part_1.csv")

# Load BERT tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

# Function to get BERT embeddings
def get_bert_embeddings(text):
    # Tokenize input text with padding, truncation, and max length set to 128
    inputs = tokenizer(text, return_tensors="pt", padding='max_length', truncation=True, max_length=128)

    # Forward pass through BERT model
    with torch.no_grad():
        outputs = model(**inputs)
        # Extract the output of the [CLS] token (first token)
        cls_embedding = outputs.last_hidden_state[:, 0, :].squeeze()

    return cls_embedding.numpy()  # Convert to numpy array for easier handling

# Apply the function to each row in the 'text' column
df['bert_embeddings'] = df['text'].apply(lambda x: get_bert_embeddings(x))

# Convert the list of arrays into a 2D numpy array
embeddings = np.stack(df['bert_embeddings'].values)

# Save embeddings to a .npy file
output_file = "/content/drive/MyDrive/MiB Dataset/2048 Each User/Splits/bert_embeddings.npy"
np.save(output_file, embeddings)

print(f"BERT embeddings saved to {output_file}")
```
