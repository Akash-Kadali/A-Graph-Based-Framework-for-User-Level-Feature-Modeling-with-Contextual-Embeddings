```python
from google.colab import drive
drive.mount('/content/drive/')

import pandas as pd
import numpy as np

import os
import shutil

# Directory containing the CSV files
source_dir = '/content/drive/MyDrive/MiB Dataset/User_Csv_Files_Val'

# Directory to move subfolders into
destination_dir = '/content/drive/MyDrive/MiB Dataset/Verify Split Files Val'

# Number of subfolders
num_subfolders = 5

# Create subfolders
for i in range(num_subfolders):
    subfolder_name = f'subfolder_{i+1}'
    os.makedirs(os.path.join(destination_dir, subfolder_name), exist_ok=True)

# Get list of CSV files
csv_files = [f for f in os.listdir(source_dir) if f.endswith('.csv')]

# Calculate number of files per subfolder
files_per_subfolder = len(csv_files) // num_subfolders
extra_files = len(csv_files) % num_subfolders

# Move files into subfolders
file_index = 0
for i in range(num_subfolders):
    subfolder_name = f'subfolder_{i+1}'
    subfolder_path = os.path.join(destination_dir, subfolder_name)

    # Determine the number of files for this subfolder
    if i < extra_files:
        num_files = files_per_subfolder + 1
    else:
        num_files = files_per_subfolder

    # Move files
    for _ in range(num_files):
        if file_index < len(csv_files):
            src_file = os.path.join(source_dir, csv_files[file_index])
            dst_file = os.path.join(subfolder_path, csv_files[file_index])
            shutil.move(src_file, dst_file)
            file_index += 1

# Count CSV files in each subfolder
for i in range(num_subfolders):
    subfolder_name = f'subfolder_{i+1}'
    subfolder_path = os.path.join(destination_dir, subfolder_name)

    # List CSV files in the subfolder
    csv_files_in_subfolder = [f for f in os.listdir(subfolder_path) if f.endswith('.csv')]

    # Count and print the number of CSV files
    count = len(csv_files_in_subfolder)
    print(f'{subfolder_name} contains {count} CSV files.')
```

```python
from google.colab import drive
drive.mount('/content/drive/')

import pandas as pd
import numpy as np
import os
import torch
from transformers import BertTokenizer, BertModel
from tqdm import tqdm

# Check if GPU is available and set the device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Load BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased').to(device)  # Move model to GPU

def get_bert_embeddings(texts):
    # Tokenize and get input IDs and attention masks
    inputs = tokenizer(texts, return_tensors='pt', truncation=True, padding=True, max_length=512)

    # Move inputs to the same device as the model
    inputs = {key: value.to(device) for key, value in inputs.items()}

    # Get embeddings from BERT
    with torch.no_grad():
        outputs = model(**inputs)

    # Extract last hidden state
    last_hidden_state = outputs.last_hidden_state
    return last_hidden_state

def max_pooling(embeddings):
    # Max pooling along the sequence dimension
    return torch.max(embeddings, dim=1).values

# Define the path to the folder containing CSV files
folder_path = '/content/drive/MyDrive/MiB Dataset/Verify Split Files Val/subfolder_1/'

# List all CSV files in the folder
files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Initialize dictionary to store user-wise pooled outputs
user_pooled_outputs = {}

# Wrap the file processing loop with tqdm to show progress for each file
for file_name in tqdm(files, desc="Processing CSV files"):
    file_path = os.path.join(folder_path, file_name)
    print(f"Processing file: {file_path}")

    try:
        # Load the CSV file
        df = pd.read_csv(file_path)

        # Extract the user ID and text columns
        user_ids = df['user_id'].tolist()  # Replace 'user_id' with the actual column name
        texts = df['text'].tolist()  # Replace 'text' with the actual column name

        # Batch processing of texts
        batch_size = 32  # Adjust the batch size based on your GPU capacity
        for i in tqdm(range(0, len(texts), batch_size), desc="Processing texts"):
            batch_texts = texts[i:i + batch_size]
            batch_user_ids = user_ids[i:i + batch_size]

            try:
                embeddings = get_bert_embeddings(batch_texts)
                pooled_output = max_pooling(embeddings)

                # Move pooled output to CPU
                pooled_output = pooled_output.cpu()

                # Add pooled output to the corresponding user's list
                for user_id, output in zip(batch_user_ids, pooled_output):
                    if user_id not in user_pooled_outputs:
                        user_pooled_outputs[user_id] = []
                    user_pooled_outputs[user_id].append(output)

            except Exception as e:
                print(f"Error processing batch starting with text: {batch_texts[0][:30]}... Error: {e}")

    except Exception as e:
        print(f"Error processing file {file_name}: {e}")

# Compute the final pooled output for each user
final_user_embeddings = []

for user_id, outputs in user_pooled_outputs.items():
    if outputs:
        # Stack and aggregate the pooled outputs for each user
        pooled_outputs_tensor = torch.stack(outputs)
        user_embedding = torch.max(pooled_outputs_tensor, dim=0).values  # Max pooling across all texts
        final_user_embeddings.append(user_embedding)

# Convert final embeddings to a tensor
if final_user_embeddings:
    final_user_embeddings_tensor = torch.stack(final_user_embeddings)
    print(f"Shape of final user embeddings: {final_user_embeddings_tensor.shape}")
else:
    print("No user embeddings to aggregate.")

# Define a directory to save the embeddings
save_folder_path = '/content/drive/MyDrive/MiB Dataset/Embeddings_Val/'
os.makedirs(save_folder_path, exist_ok=True)  # Create the directory if it doesn't exist

# Save the final user embeddings to a file
if final_user_embeddings_tensor is not None:
    # Convert tensor to NumPy array before saving
    final_user_embeddings_numpy = final_user_embeddings_tensor.cpu().numpy()

    # Save as a NumPy array (.npy)
    save_path_npy = os.path.join(save_folder_path, 'final_user_embeddings.npy')
    np.save(save_path_npy, final_user_embeddings_numpy)  # Save the embeddings
    print(f"Embeddings saved to: {save_path_npy}")

else:
    print("No user embeddings to save.")
```

```python
from google.colab import drive
drive.mount('/content/drive/')

import pandas as pd
import numpy as np
import os
import torch
from transformers import BertTokenizer, BertModel
from tqdm import tqdm

# Check if GPU is available and set the device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Load BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased').to(device)  # Move model to GPU

def get_bert_embeddings(texts):
    # Tokenize and get input IDs and attention masks
    inputs = tokenizer(texts, return_tensors='pt', truncation=True, padding=True, max_length=512)

    # Move inputs to the same device as the model
    inputs = {key: value.to(device) for key, value in inputs.items()}

    # Get embeddings from BERT
    with torch.no_grad():
        outputs = model(**inputs)

    # Extract last hidden state
    last_hidden_state = outputs.last_hidden_state
    return last_hidden_state

def max_pooling(embeddings):
    # Max pooling along the sequence dimension
    return torch.max(embeddings, dim=1).values

# Define the path to the folder containing CSV files
folder_path = '/content/drive/MyDrive/MiB Dataset/Verify Split Files Val/subfolder_5/'

# List all CSV files in the folder
files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Initialize dictionary to store user-wise pooled outputs
user_pooled_outputs = {}

# Wrap the file processing loop with tqdm to show progress for each file
for file_name in tqdm(files, desc="Processing CSV files"):
    file_path = os.path.join(folder_path, file_name)
    print(f"Processing file: {file_path}")

    try:
        # Load the CSV file
        df = pd.read_csv(file_path)

        # Extract the user ID and text columns
        user_ids = df['user_id'].tolist()  # Replace 'user_id' with the actual column name
        texts = df['text'].tolist()  # Replace 'text' with the actual column name

        # Batch processing of texts
        batch_size = 32  # Adjust the batch size based on your GPU capacity
        for i in tqdm(range(0, len(texts), batch_size), desc="Processing texts"):
            batch_texts = texts[i:i + batch_size]
            batch_user_ids = user_ids[i:i + batch_size]

            try:
                embeddings = get_bert_embeddings(batch_texts)
                pooled_output = max_pooling(embeddings)

                # Move pooled output to CPU
                pooled_output = pooled_output.cpu()

                # Add pooled output to the corresponding user's list
                for user_id, output in zip(batch_user_ids, pooled_output):
                    if user_id not in user_pooled_outputs:
                        user_pooled_outputs[user_id] = []
                    user_pooled_outputs[user_id].append(output)

            except Exception as e:
                print(f"Error processing batch starting with text: {batch_texts[0][:30]}... Error: {e}")

    except Exception as e:
        print(f"Error processing file {file_name}: {e}")

# Compute the final pooled output for each user
final_user_embeddings = []

for user_id, outputs in user_pooled_outputs.items():
    if outputs:
        # Stack and aggregate the pooled outputs for each user
        pooled_outputs_tensor = torch.stack(outputs)
        user_embedding = torch.max(pooled_outputs_tensor, dim=0).values  # Max pooling across all texts
        final_user_embeddings.append(user_embedding)

# Convert final embeddings to a tensor
if final_user_embeddings:
    final_user_embeddings_tensor = torch.stack(final_user_embeddings)
    print(f"Shape of final user embeddings: {final_user_embeddings_tensor.shape}")
else:
    print("No user embeddings to aggregate.")

# Define a directory to save the embeddings
save_folder_path = '/content/drive/MyDrive/MiB Dataset/Embeddings_Val/'
os.makedirs(save_folder_path, exist_ok=True)  # Create the directory if it doesn't exist

# Save the final user embeddings to a file
if final_user_embeddings_tensor is not None:
    # Convert tensor to NumPy array before saving
    final_user_embeddings_numpy = final_user_embeddings_tensor.cpu().numpy()

    # Save as a NumPy array (.npy)
    save_path_npy = os.path.join(save_folder_path, 'final_user_embeddings5.npy')
    np.save(save_path_npy, final_user_embeddings_numpy)  # Save the embeddings
    print(f"Embeddings saved to: {save_path_npy}")

else:
    print("No user embeddings to save.")
```
