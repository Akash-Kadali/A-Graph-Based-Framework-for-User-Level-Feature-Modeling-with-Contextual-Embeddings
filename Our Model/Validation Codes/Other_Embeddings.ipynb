```python
import pandas as pd
import numpy as np
from tqdm import tqdm

# Path to the train.csv file
csv_file_path = '/content/drive/MyDrive/MiB Dataset/A Fresh Start/User/train.csv'

# Load the CSV file
df = pd.read_csv(csv_file_path)

# Extract required columns
required_columns = ['user_id', 'statuses_count', 'followers_count', 'friends_count', 'favourites_count', 'type']
df = df[required_columns]

# Initialize lists to store features and user IDs
user_id_list = []
statuses_count_list = []
followers_count_list = []
friends_count_list = []
favourites_count_list = []
type_list = []

# Group by user_id and extract features
for user_id, group in tqdm(df.groupby('user_id'), desc="Processing User Features"):
    try:
        # Extract user-specific features (assume consistent values within groups)
        statuses_count = group['statuses_count'].iloc[0]
        followers_count = group['followers_count'].iloc[0]
        friends_count = group['friends_count'].iloc[0]
        favourites_count = group['favourites_count'].iloc[0]
        user_type = group['type'].iloc[0]

        # Append user_id and features to respective lists
        user_id_list.append(user_id)
        statuses_count_list.append(statuses_count)
        followers_count_list.append(followers_count)
        friends_count_list.append(friends_count)
        favourites_count_list.append(favourites_count)
        type_list.append(user_type)

    except Exception as e:
        print(f"Error processing user {user_id}: {e}")

# Save each feature list and user_id as separate .npy files
np.save('/content/drive/MyDrive/MiB Dataset/A Fresh Start/User/user_id.npy', np.array(user_id_list))
np.save('/content/drive/MyDrive/MiB Dataset/A Fresh Start/User/statuses_count.npy', np.array(statuses_count_list))
np.save('/content/drive/MyDrive/MiB Dataset/A Fresh Start/User/followers_count.npy', np.array(followers_count_list))
np.save('/content/drive/MyDrive/MiB Dataset/A Fresh Start/User/friends_count.npy', np.array(friends_count_list))
np.save('/content/drive/MyDrive/MiB Dataset/A Fresh Start/User/favourites_count.npy', np.array(favourites_count_list))
np.save('/content/drive/MyDrive/MiB Dataset/A Fresh Start/User/type.npy', np.array(type_list))

print("User IDs and features saved successfully as separate .npy files.")
```
